## 2. Data Manipulation

##### Question 2a: Create a new column -’total_benefitted’ that is a sum of inmates benefitted through all modes.[Hint: Use sum method with appropriate axis]

```python
# Creating the new column 'total_benefitted’ as sum of all the other modes of benefitting the inmates
df['total_benefitted'] = df['No. of Inmates benefitted by Elementary Education']+ df['No. of Inmates benefitted by Adult Education']+df['No. of Inmates benefitted by Higher Education']+df['No. of Inmates benefitted by Computer Course']
df.head()
```

#### Question 2b: Create a new row -“totals” that is the sum of all inmates benefitted through each mode across all states.

```python
# Creating new row 'totals' that is the sum of all inmates benefitted through each mode across all states.
df.loc['Total',2:]= df.sum(axis=0)
df
```

## 3. Plotting

#### Question 3a: Make a bar plot with each state name on the x -axis and their total benefitted inmates astheir bar heights. Which state has the maximum number of beneficiaries?

```python
fig,ax=plt.subplots(figsize=(60,20))
ax = sns.barplot(x="STATE/UT", y="total_benefitted", data=df)
ax.set(xlabel = 'STATE/UT', ylabel = 'Total Benefitted');
```

It is clear that Andhra Pradesh has maximum number of beneficiaries followed by Uttar Pradesh.

#### Question 3b. Make a pie chart that depicts the ratio among different modes of benefits

```python
## Taking the data of modes of benefit columns and find their ratio  
data = [37027, 53720, 8311, 7356]
values = []
for data in data:
    value= (data/106414)
    values.append(value)
print(values)
```

```python
plt.subplots(figsize=(10,10))
benefit_modes= ['No. of Inmates benefitted by Elementary Education','No. of Inmates benefitted by Adult Education','No. of Inmates benefitted by Higher Education','No. of Inmates benefitted by Computer Course']

plt.pie(values, labels = benefit_modes,autopct='%1.2f%%')
plt.show()
```
