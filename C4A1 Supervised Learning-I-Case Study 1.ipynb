#### Question 1: We will use acoustic features to distinguish a male voice from female. Load the dataset from “voice.csv”,identify the target variable and do a one-hot encoding for the same. Split the dataset in train-test with 20% of the data kept aside for testing.[Hint: Refer to LabelEncoder documentation in scikit-learn

```python
# Importing Label Encoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['label']= le.fit_transform(df['label'])
```

```python
df.head(5)
```

```python
# Importing train_test_split
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('label', axis=1)
y=df['label']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

#### Question 2.Fit a logistic regression model and measure the accuracy on the test set.[Hint:Refer to Linear Models section in scikit-learn]

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_train,y_train)
```

```python
y_preds=lr.predict(X_test)
```

```python
lr.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_preds))
```

### The accuracy of the model comes out to be 0.90
