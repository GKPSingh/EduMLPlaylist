```python
# Scalling the data
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df[['Avg_Session_Length']])
df['Avg_Session_Length']=scaler.transform(df[['Avg_Session_Length']])
```

```python
# Scalling the Time_on_App column
scaler.fit(df[['Time_on_App']])
df['Time_on_App']=scaler.transform(df[['Time_on_App']])
```

```python
# Scalling the Time_on_Website column
scaler.fit(df[['Time_on_Website']])
df['Time_on_Website']=scaler.transform(df[['Time_on_Website']])
```

```python
# Scalling the Length_of_Membership column
scaler.fit(df[['Length_of_Membership']])
df['Length_of_Membership']=scaler.transform(df[['Length_of_Membership']])
```

```python
# Checking the data
df.head()
```

```python
# Dropping the Email and Address Columns
df.drop(['Email', 'Address'], axis=1, inplace=True)
df.head()
```

```python
# Correlation Plot
df_corr = df.corr()
fig,ax = plt.subplots(figsize=(10,8))
ax = sns.heatmap(df_corr,annot=True,fmt='.2f',lw=1, cmap='Blues')
```

From correlation heatmp, it is clear that Length_of_Membership, Time_on_App, and Avg_Session_Length are most coorelated with the Yearly_Amount_Spent

```python
# Importing Train_test_split
from sklearn.model_selection import train_test_split
```

```python
# Splitting the data
X= df.drop('Yearly_Amount_Spent', axis=1)
y= df['Yearly_Amount_Spent']
```

### Question5: Compute Train and Test the data and answer multiple questions --What is the use of random_state=85?

```python
# Splitting the data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=85)
```

The use of random_state will give us the same results every time.

```python
# Importing the model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# Fitting the data
lr.fit(X_train,y_train)
lr.score(X_train,y_train)
```

```python
# Making the prediction
y_preds= lr.predict(X_test)
```

```python
# Checking Predicitions
y_preds
```

```python
# Checking the test data score
lr.score(X_test,y_test)
```

### Question6: Compute Predict the data and do a scatter plot. Check if actual and predicted data match?

```python
plt.scatter(y_test, y_preds)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

Actual and predicted data is matching with little variation

```python
# Importing the mean_absolute_error and mean_squared_error
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Finding mean_absolute_error
mae=print(mean_absolute_error(y_test,y_preds))
mae
```

```python
# Importing  mean_squared_error
mse = print(mean_squared_error(y_test,y_preds))
mse
```

### Question7: What is the value of Root Mean Squared Error?

```python
rmse = print(np.sqrt(mean_squared_error(y_test,y_preds)))
rmse
```

### Question8: Final Question Based on coefficients interpret company should focus more on their mobile app or on their website

```python
df = pd.DataFrame(data = lr.coef_, columns = ['Coefficient'],index = X_train.columns)
df.head()
```

From the findings, it is clear that Time_on_Website is not that significant in generating revenues but Time_on_App has been very significant in revenue generation. Still, it does not mean that company should scrap the website. Instead, company must work to improve the website so that more customers can be attacked to the website. It is also notable that revenue generation is highly dependent on Length_of_Membership.
