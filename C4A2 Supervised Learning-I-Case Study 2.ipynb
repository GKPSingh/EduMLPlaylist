### Question 2: This dataset contains many categorical features, replace them with label encoding.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cols_list = ['surgery','age','temp_of_extremities','peripheral_pulse','mucous_membrane','capillary_refill_time','pain','peristalsis','abdominal_distention','nasogastric_tube','nasogastric_reflux','rectal_exam_feces','abdomen','abdomo_appearance','outcome','surgical_lesion','cp_data']
def create_labelencoder(cols_list):
    '''
    Creates Label Encoders and saves them as numpy files in (Root + LabelEncoder) path
    
    Args:
        List of dataframe columns
    '''
    for col in cols_list:
        df[col]= df[col].astype('str')
        le.fit(df[col])
        df[col] = le.transform(df[col])
        
        
create_labelencoder(cols_list)
```

### Quesion 3: Replace the missing values by the most frequent value in each column.

```python
rectal_temp_median = df['rectal_temp'].median()
df['rectal_temp'].fillna(rectal_temp_median, inplace = True)

pulse_median = df['pulse'].median()
df['pulse'].fillna(pulse_median, inplace = True)

respiratory_rate_median = df['respiratory_rate'].median()
df['respiratory_rate'].fillna(respiratory_rate_median, inplace = True)

nasogastric_reflux_ph_median = df['nasogastric_reflux_ph'].median()
df['nasogastric_reflux_ph'].fillna(nasogastric_reflux_ph_median, inplace = True)

packed_cell_volume_median = df['packed_cell_volume'].median()
df['packed_cell_volume'].fillna(packed_cell_volume_median, inplace = True)

total_protein_median = df['total_protein'].median()
df['total_protein'].fillna(total_protein_median, inplace = True)

abdomo_protein_median = df['abdomo_protein'].median()
df['abdomo_protein'].fillna(abdomo_protein_median, inplace = True)

                              
```

```python
X = df.drop('outcome', axis=1)
y = df['outcome']
```

```python
from sklearn.model_selection import train_test_split
```

### Question 4: Fit a decision tree classifier and observe the accuracy

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(X_train,y_train)
```

```python
dtc.score(X_train,y_train)
```

```python
y_preds = dtc.predict(X_test)
```

```python
dtc.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_preds))
```

### The accuracy of the model using decision tree classifier is 66.6%
